```python
import os
import pandas as pd
from yolov8_infer import model

COCO_ROOT = "C:/Users/phiho/Projects/Dataset/MS-COCO/COCO_81/coco81/train"
data = []
min_num_imgs = 1000000

for subfolder_name in os.listdir(COCO_ROOT):
    subfolder_path = os.path.join(COCO_ROOT, subfolder_name)
    
    if os.path.isdir(subfolder_path):
        jpg_count = 0
        correct_label = 0
        for filename in os.listdir(subfolder_path):
            label = model(os.path.join(subfolder_path, filename))
            if label == get_ground_truth(filename): 
                COCO_TRAIN_LABELS_PREDICT_TRUE[subfolder_path.split("\\")[-1]].append(filename)
                correct_label += 1
            else:                
                COCO_TRAIN_LABELS_PREDICT_FALSE[subfolder_path.split("\\")[-1]].append(filename)
            if filename.lower().endswith('.jpg'):
                jpg_count += 1

        data.append([subfolder_path.split("\\")[-1], jpg_count, f"{correct_label/jpg_count*100:.2f}%"])
        min_num_imgs = min(min_num_imgs, jpg_count)


df = pd.DataFrame(data, columns=['Class', 'Number of all images', 'Raw accuracy'])
# df = pd.DataFrame(data, columns=['Class', 'Number of all images', 'Raw accuracy', 'Number of training images', 'Number of testing images', 'Accuracy of training images'])
df.to_excel('C:/Users/phiho/Projects/GAN_Yolov8/analysis/Data.xlsx', sheet_name='Sheet1', index=False)
```

```python
import os
import random
import shutil
import pandas as pd

COCO_ACC96_ROOT_TRAIN = "C:/Users/phiho/Projects/Dataset/MS-COCO/COCO_96%_YOLOV8/train"
COCO_ACC96_ROOT_TEST = "C:/Users/phiho/Projects/Dataset/MS-COCO/COCO_96%_YOLOV8/test"

for label_name in COCO_CLASS_NAMES:
    subfolder_path = os.path.join(COCO_ROOT, label_name)
    desfolder_path = os.path.join(COCO_ACC96_ROOT_TRAIN, label_name)
    test_path = os.path.join(COCO_ACC96_ROOT_TEST, label_name)
    if not os.path.exists(desfolder_path):
        os.makedirs(desfolder_path)
    if not os.path.exists(test_path):
        os.makedirs(test_path)
    
    # File true
    files_true = [f for f in COCO_TRAIN_LABELS_PREDICT_TRUE[label_name] if os.path.isfile(os.path.join(subfolder_path, f))]
    num_files_true_to_move = len(files_true)
    if label_name == "person":
        num_files_true_to_move = 5000

    files_true_to_move = random.sample(files_true, num_files_true_to_move)
    
    for file in files_true_to_move:
        src_path = os.path.join(subfolder_path, file)
        dst_path = os.path.join(desfolder_path, file)
        shutil.copy(src_path, dst_path)

    # File false
    files_false = [f for f in COCO_TRAIN_LABELS_PREDICT_FALSE[label_name] if os.path.isfile(os.path.join(subfolder_path, f))]
    num_files_false_to_move = int(num_files_true_to_move/96*4) + 1

    files_false_to_move = random.sample(files_false, num_files_false_to_move)
    
    for file in files_false_to_move:
        src_path = os.path.join(subfolder_path, file)
        dst_path = os.path.join(desfolder_path, file)
        shutil.copy(src_path, dst_path)

    # Merge both true and false
    files_both = files_true_to_move + files_false_to_move

    # File test
    if len(COCO_TRAIN_LABELS_PREDICT_TRUE[label_name]) > 4:
        num_files_test_to_move = int(len(files_both) * 0.2) + 1
        num_files_train_to_move = len(files_both) - num_files_test_to_move
        files_test_to_move = random.sample(files_both, num_files_test_to_move)
        
        num_files_test_true = 0
        num_files_test_false = 0
        for file in files_test_to_move:
            if file in COCO_TRAIN_LABELS_PREDICT_TRUE[label_name]: num_files_test_true += 1
            elif file in COCO_TRAIN_LABELS_PREDICT_FALSE[label_name]: num_files_test_false += 1
            else: 
                print(label_name, file)
                print("Error")
            src_path = os.path.join(desfolder_path, file)
            dst_path = os.path.join(test_path, file)
            shutil.move(src_path, dst_path)
    else:
        num_files_test_to_move = int(len(files_both) * 0.2) + 1
        num_files_train_to_move = len(files_both) - num_files_test_to_move
        files_test_to_move = random.sample(files_both, num_files_test_to_move)
        
        num_files_test_true = 0
        num_files_test_false = 0
        for file in files_test_to_move:
            if file in COCO_TRAIN_LABELS_PREDICT_TRUE[label_name]: num_files_test_true += 1
            elif file in COCO_TRAIN_LABELS_PREDICT_FALSE[label_name]: num_files_test_false += 1
            else: 
                print(label_name, file)
                print("Error")
            src_path = os.path.join(desfolder_path, file)
            dst_path = os.path.join(test_path, file)
            shutil.move(src_path, dst_path)
    

    num_files_train_true = num_files_true_to_move - num_files_test_true
    num_files_train_false = num_files_false_to_move - num_files_test_false

    # Save to data frame
    for i in range(len(data)):
        if data[i][0] == label_name:
            data[i].append(num_files_true_to_move + num_files_false_to_move)
            data[i].append(f"{num_files_true_to_move/len(files_both)*100:.2f}%")

            sum_process_train = num_files_train_true + num_files_train_false
            if sum_process_train == 0: sum_process_train = 1

            sum_process_test = num_files_test_true + num_files_test_false
            if sum_process_test == 0: sum_process_test = 1

            data[i].append(num_files_train_to_move)
            data[i].append(f"{num_files_train_true/sum_process_train*100:.2f}%")
            data[i].append(num_files_test_to_move)
            data[i].append(f"{num_files_test_true/sum_process_test*100:.2f}%")
            break
        
```
